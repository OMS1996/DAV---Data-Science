# 1. Introduction

The Big Five Model, also known as the Five-Factor Model, is the most widely accepted personality theory held by psychologists today. The theory states that personality can be boiled down to five core factors, known by the acronym CANOE or OCEAN.

1. Conscientiousness
    - Conscientiousness describes a person’s ability to regulate their impulse control in order to engage in goal-directed behaviors (Grohol, 2019). It measures elements such as control, inhibition, and persistency of behavior.
2. Agreeableness
    - Agreeableness refers to how people tend to treat relationships with others. Unlike extraversion which consists of the pursuit of relationships, agreeableness focuses on people’s orientation and interactions with others (Ackerman, 2017).
3. Neuroticism
    - Neuroticism describes the overall emotional stability of an individual through how they perceive the world. It considers how likely a person is to interpret events as threatening or difficult. (John & Srivastava, 1999)
4. Openness to Experience
    - Openness to experience refers to one’s willingness to try new things as well as engage in imaginative and intellectual activities. It includes the ability to “think outside of the box.” (John & Srivastava, 1999)
5. Extraversion
    - Extraversion reflects the tendency and intensity to which someone seeks interaction with their environment, particularly socially. It encompasses the comfort and assertiveness levels of people in social situations. Additionally, it also reflects the sources from which someone draws energy. (John & Srivastava, 1999):

### 5.3 Machine Learning

#### 5.3.0 The approach from a bird's eye view.
In order to achieve our goals, we have to use a blend of unsupervised learning and supervised learning. first we will use an unsupervised clustering algorithm to group the data together into clusters that represent different personalities the quality of the clustering algorithm (K-means / Hierarchical clustering) will be measured using inertia and silhouette cluster performance metrics, then we move on to creating several supervised learning models from which we will choose the top performing ones using our evaluation metrics used for classification. finally, we would combine the top performing models into a single ensemble method which would perform a classification based on the average of the probabilities across all classes. The next subsections explain the approach in more detail.

#### 5.3.1 Unsupervised learning
In this project we will be using either Hierarchical clustering or K-means, how we decide is going to be through our understanding of the dataset, domain knowledge and probably through an empirical approach, however whether we choose K-means or Hierarchical this would be how we approach both of them.

- K - Means:
we start starts with a first group of randomly selected centroids, which are used as the beginning points for every cluster, and then performs iterative (repetitive) calculations to optimize the positions of the centroids
Finally It halts creating and optimizing clusters when either:
The centroids have stabilized — there is no change in their values because the clustering has been successful.
The defined number of iterations has been achieved. How we evaluate the best number of K will be through the performance metrics "inertia" and/or "silhouettes"

- Hierarchical clustering:
    - Make each data point a single-point cluster → forms N clusters
    - Take the two closest data points and make them one cluster → forms N-1 clusters
    - Take the two closest clusters and make them one cluster → Forms N-2 clusters.
    - Repeat step-3 until you are left with only one cluster.
    
By the end of this part we will have N clusters representing N different personalities. (N refers to the number of clusters)

#### 5.3.2 Supervised learning
After clustering we need to create predictive supervised learning model(s) to help classify the data into of the N clusters previously mentioned in section `5.3.1`
For our supervised learning models we will create 3 types of models a neural network, LightGBM and random forests/SVMs with potentially different versions for each one of the mentioned models and then choose amongst them the top performing ones relative to our classification evaluation metrics (examples: accuracy , precision , recall ,AUC....etc.).


#### 5.3.3 Ensemble 
The methodology we will use for this classification problem is that we will consider the probabilities of each classes returned by all the models, average these probabilities and keep the class with the highest average probability (soft-voting) which will decide which class it should be or rather which personality type it would be.
